<p style='font-size:20px;font-family:Arial;color:#00233C'><b>Introduction:</b></p>
<p style='font-size:16px;font-family:Arial;color:#00233C'>This notebook describes how to get started with the Azure-OpenAI Service and provides step-by-step instructions to create a resource and deploy a model.</p>

<hr style='height:2px;border:none;background-color:#00233C;'>
<b style='font-size:20px;font-family:Arial;color:#00233C'>1. Create a resource</b>

<ol style='font-size:16px;font-family:Arial;color:#00233C'>
    <li>Sign in with your Azure subscription in the Azure portal.</li>
    <li>Search for the Azure OpenAI.</li>
    <li>Select Create Azure OpenAI.</li>
</ol>

<img src='./images/step-1.png' alt='Step 1'>

<p style='font-size:16px;font-family:Arial;color:#00233C'>On the Create Azure OpenAI page, provide the following information for the fields on the Basics tab:</p>

<ul style='font-size:16px;font-family:Arial;color:#00233C'>
    <li><b>Subscription</b>: The Azure subscription used in your Azure OpenAI Service onboarding application.</li>
    <li><b>Resource group</b>: The Azure resource group to contain your Azure OpenAI resource. You can create a new group or use a pre-existing group.</li>
    <li><b>Region</b>: The location of your instance. Different locations can introduce latency, but they don't affect the runtime availability of your resource.</li>
    <li><b>Name</b>: A descriptive name for your Azure OpenAI Service resource, such as MyOpenAIResource.</li>
    <li><b>Pricing Tier</b>: The pricing tier for the resource. Currently, only the Standard tier is available for the Azure OpenAI Service. For more info on pricing, visit the <a href='https://azure.microsoft.com/pricing/details/cognitive-services/openai-service/' style='color:#0066CC;text-decoration:none;'>Azure OpenAI pricing page</a>.</li>
    <li><b>Select Next</b>.</li>
</ul>

<img src='./images/step-2.png' alt='Step 2'>

<hr style='height:2px;border:none;background-color:#00233C;'>
<b style='font-size:20px;font-family:Arial;color:#00233C'>2. Configure network security</b>

<p style='font-size:16px;font-family:Arial;color:#00233C'>The Network tab presents three options for the security type:</p>

<ul style='font-size:16px;font-family:Arial;color:#00233C'>
    <li><b>Option 1</b>: All networks, including the internet, can access this resource.</li>
    <li><b>Option 2</b>: Selected networks, configure network security for your Azure AI services resource.</li>
    <li><b>Option 3</b>: Disabled, no networks can access this resource. You can configure private endpoint connections that will be the exclusive way to access this resource.</li>
    <li><b>Select Next</b>: Here we go with Option 1.</li>
</ul>

<img src='./images/step-3.png' alt='Step 3'>

<hr style='height:2px;border:none;background-color:#00233C;'>
<b style='font-size:20px;font-family:Arial;color:#00233C'>3. Confirm the configuration and create the resource</b>

<p style='font-size:16px;font-family:Arial;color:#00233C'>Configure any tags for your resource, as desired.</p>

<img src='./images/step-4.png' alt='Step 4'>

<p style='font-size:16px;font-family:Arial;color:#00233C'>Select Next to move to the final stage in the process: <b>Submit</b>.</p>
<p style='font-size:16px;font-family:Arial;color:#00233C'>Confirm your configuration settings, and select <b>Create</b>.</p>

<img src='./images/step-5.png' alt='Step 5'>

<img src='./images/step-6.png' alt='Step 6'>

<p style='font-size:16px;font-family:Arial;color:#00233C'>The Azure portal displays a notification when the new resource is available. Select <b>Go to resource</b>.</p>

<img src='./images/step-7.png' alt='Step 7'>

<hr style='height:2px;border:none;background-color:#00233C;'>
<b style='font-size:20px;font-family:Arial;color:#00233C'>4. Deploy Model</b>

<p style='font-size:16px;font-family:Arial;color:#00233C'>Open the Model Catalog to see models available for you to deploy.</p>

<img src='./images/step-8.png' alt='Step 8'>

<p style='font-size:16px;font-family:Arial;color:#00233C'>Model availability varies by region. For a list of available models per region, see the <a href='https://learn.microsoft.com/en-us/azure/ai-services/openai/concepts/models#model-summary-table-and-region-availability' style='color:#0066CC;text-decoration:none;'>Model summary table and region availability</a>. Select the model of your choice.</p>

<img src='./images/step-9.png' alt='Step 9'>

<p style='font-size:16px;font-family:Arial;color:#00233C'>Here we select the <b>gpt-4o-mini</b> model.</p>

<img src='./images/step-10.png' alt='Step 10'>

<p style='font-size:16px;font-family:Arial;color:#00233C'>The deployment tab presents options for the deployment configuration:</p>

<ul style='font-size:16px;font-family:Arial;color:#00233C'>
    <li><b>Deployment name</b>: Choose a name carefully. The deployment name is used in your code to call the model using the client libraries and the REST APIs.</li>
    <li><b>Deployment type</b>: Standard, Global-Batch, Global-Standard, Provisioned-Managed. Learn more about <a href='https://learn.microsoft.com/en-us/azure/ai-services/openai/how-to/deployment-types' style='color:#0066CC;text-decoration:none;'>deployment type options</a>.</li>
    <li><b>Tokens</b>: For the Tokens per Minute Rate Limit, adjust the Tokens per Minute (TPM) to set the effective rate limit for your deployment.</li>
    <li><b>Select Deploy</b>.</li>
</ul>

<img src='./images/step-11.png' alt='Step 11'>

<p style='font-size:16px;font-family:Arial;color:#00233C'>The deployments table shows a new entry that corresponds to your newly created model.<br><br>When the deployment completes, your model deployment status changes to succeeded.</p>

<img src='./images/step-12.png' alt='Step 12'>
