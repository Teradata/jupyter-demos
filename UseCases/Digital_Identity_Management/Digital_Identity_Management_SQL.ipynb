<p style = 'font-size:16px;font-family:Arial'> Let's look at the data on how we can get the answers to our questions once we have implemented the DIM solution.    

<p style = 'font-size:16px;font-family:Arial'>For the demo, we are using the follwing tables populated via Celebrus.

- **CampaignAttributionEnd**: Populated when session ends, summarises information about the session in total
- **CampaignAttributionStart**: Populated when visits begin, identifies the campaigns that brought the visitor into a visit during a session
- **Click**: Identifies a click on any object, plus information about that object
- **DataLayer**: Demonstrates we can capture data layers. For the XYZ POC we captured the Tealium data layer
- **DeviceStatus**: Flags when multiple IP addresses are being used by a device
- **FieldInteractionIdentification**: Identifies a fields being interacted with and Details of the interaction with a field, including what's input, and how many times
- **IndividualAttribute**: Identifies attributes about the Individual - for the POC we have collected Adobe Marketing Cloud IDs
- **Page**: Populated when a Page loads, contains details about the page, plus where in the session the page is visited
- **PageSummary**: Populated when the page unloads, summarises timings on page, scrolling and other key stats
- **SessionStart**: Session Start events generated at the beginning of a session. Creates a Session Key and identifies previous sessions for the device being used
- **SessionSummary**: Session summary generated at the end of each session, summarises timings, page counts and other key statistics
- **SignalandStrength**: Generated when NLP/ML algorithm detects engagement with products and services and scores this. Tunable and extensible
- **SystemIdentityAttributes**: Attributes captured to define identity. For the POC we included the CelebrusCookie and CUST ID
- **TransactionComplete**: Configurable: For the POC we configured the XYZ Loan Calculator (Public Site) - Captures completed loan calculator interactions (last one) - useful in personalisation
- **TransactionStep**: Configurable: For the POC we configured the XYZ Loan Calculator (Public Site) - Captures individual steps when a customer runs repeated scenarios
- **Visitor**: Writes a record every time Identity changes during visits

<h2 style = 'font-size:20px;font-family:Arial'><b>Find customers in multiple domains i.e same customer is logged in multiple domains</b></h2>
<p style = 'font-size:16px;font-family:Arial'>For this we look at the Visitor table for a session.

```
select * from demo_user.visitor where sessionkey = '46421275-a136-4935-84fa-8683c1d105cc' order by eventtimestamp;
```

<p style = 'font-size:16px;font-family:Arial'>From the above query we can see that when the customer logged in the Customerid is populated but the SessionKey and CookieUniqueVisitorTrackingId remains the same.

```
select * from demo_user.page where sessionkey = '46421275-a136-4935-84fa-8683c1d105cc' order by eventtimestamp;
```

<p style = 'font-size:16px;font-family:Arial'>From the above output we can see that in the same session (SeessionKey) we have multiple domains (PageDomainLocation). We can also check how many different domains customer browsed in each session. 

```
select sessionkey, count(distinct pagelocationdomain) from demo_user.page group by 1;
```

<h2 style = 'font-size:20px;font-family:Arial'><b>Find the multi-user device</b></h2>
<p style = 'font-size:16px;font-family:Arial'>i.e find the customers which are using the same device. For this we look at the Visitor table for a session.

```
select * from demo_user.visitor where sessionkey = '82d9dba0-dc35-4611-b5b4-fc41e1ad3a86' order by eventtimestamp;
```

<p style = 'font-size:16px;font-family:Arial'>From the query above we can see multiple Customerid, CustomerMultiUser associated with the same SessionKey. We can then put this SessionKey in the Page Table to get the journey for each customer.

```
select * from demo_user.page where sessionkey = '82d9dba0-dc35-4611-b5b4-fc41e1ad3a86' order by eventtimestamp;
```
