<p style = 'font-size:16px;font-family:Arial;color:#00233C'> Let's look at the data on how we can get the answers to our questions once we have implemented the DIM solution.    </p>

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>For the demo, we are using the following tables populated via Celebrus.</p>

- **CampaignAttributionEnd**: Populated when session ends, summarises information about the session in total
- **CampaignAttributionStart**: Populated when visits begin, identifies the campaigns that brought the visitor into a visit during a session
- **Click**: Identifies a click on any object, plus information about that object
- **DataLayer**: Demonstrates we can capture data layers. For the XYZ POC we captured the Tealium data layer
- **DeviceStatus**: Flags when multiple IP addresses are being used by a device
- **FieldInteractionIdentification**: Identifies a fields being interacted with and Details of the interaction with a field, including what's input, and how many times
- **IndividualAttribute**: Identifies attributes about the Individual - for the POC we have collected Adobe Marketing Cloud IDs
- **Page**: Populated when a Page loads, contains details about the page, plus where in the session the page is visited
- **PageSummary**: Populated when the page unloads, summarises timings on page, scrolling and other key stats
- **SessionStart**: Session Start events generated at the beginning of a session. Creates a Session Key and identifies previous sessions for the device being used
- **SessionSummary**: Session summary generated at the end of each session, summarises timings, page counts and other key statistics
- **SignalandStrength**: Generated when NLP/ML algorithm detects engagement with products and services and scores this. Tunable and extensible
- **SystemIdentityAttributes**: Attributes captured to define identity. For the POC we included the CelebrusCookie and CUST ID
- **TransactionComplete**: Configurable: For the POC we configured the XYZ Loan Calculator (Public Site) - Captures completed loan calculator interactions (last one) - useful in personalisation
- **TransactionStep**: Configurable: For the POC we configured the XYZ Loan Calculator (Public Site) - Captures individual steps when a customer runs repeated scenarios
- **Visitor**: Writes a record every time Identity changes during visits

<p style = 'font-size:20px;font-family:Arial'><b>Find customers in multiple domains i.e same customer is logged in multiple domains</b></p>

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>From the above table descriptions we can see that <b>Page</b> table stores the page visited in a session. Let's start by checking the data in that table.</p>

```
select sessionkey, count(distinct pagelocationdomain) from demo_dim.page group by 1 order by 2 desc;
```

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>Now let us take one Sessionkey and see how the data in the table looks like. Above, the session key '46421275-a136-4935-84fa-8683c1d105cc' had 9 distinct PageLocationDomain values.  If we look at that one session ordered by eventtimestamp, we can see the time ordered sequence of the 32 pages visited. (<i><b>Note:</b> You may need to adjust the column widths to see content)</i> </p>

```
select * from demo_dim.page where sessionkey = '46421275-a136-4935-84fa-8683c1d105cc' order by eventtimestamp;
```

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>From the above output we can see: 
<ul style = 'font-size:16px;font-family:Arial;color:#00233C'>
    <li>that in the same session (SessionKey) we have multiple domains (PageDomainLocation)</li>
    <li>in the session when the customer logs in the customerid updates from NULL to id but the Sessionkey remains the same.  <i>(Scroll to the right of the table above)</i></li>
</ul>
<p style = 'font-size:16px;font-family:Arial;color:#00233C'>
This tracking is built into Celebrus. Recall that the Visitor table writes a record every time Identity changes during visits. We can check that for the above sessionkey.</p>

```
select * from demo_dim.visitor where sessionkey = '46421275-a136-4935-84fa-8683c1d105cc' order by eventtimestamp;
```

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>From the above query we can see that when the customer logged in the Customerid is populated but the SessionKey and CookieUniqueVisitorTrackingId remains the same.</p>

<p style = 'font-size:20px;font-family:Arial'><b>Find the multi-user device</b></p>
<p style = 'font-size:16px;font-family:Arial;color:#00233C'>In cases where there are devices that can be used by multiple customers, we can still distinguish the unique customers.  We will start by looking for multiple Customerid values for the same SessionKey on the page table.</p>

```
select sessionkey, count(distinct customerid) from demo_dim.page where customerid is not null group by 1 order by 2 desc;
```

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>We see that SessionKey '82d9dba0-dc35-4611-b5b4-fc41e1ad3a86' has 2 Customerid values. <i>(<b>Note:</b> You may need to adjust the column widths to see content.)</i> Let us look at the time ordered page events including the Customerid at the right side of the table below. </p>

```
select * from demo_dim.page where sessionkey = '82d9dba0-dc35-4611-b5b4-fc41e1ad3a86' order by eventtimestamp;
```

<p style = 'font-size:16px;font-family:Arial;color:#00233C'> From the above output, <i>(scroll to right of table above)</i> we can see the customer activity in the session. We can also look at the Visitor table for the same Sessionkey to see the two non-null Customerid values from the same SessionKey.</p>

```
select * from demo_dim.visitor where sessionkey = '82d9dba0-dc35-4611-b5b4-fc41e1ad3a86' order by eventtimestamp;
```

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>From the query above we can see multiple Customerid, CustomerMultiUser associated with the same SessionKey.</p>

<p style = 'font-size:20px;font-family:Arial'><b>Cross domain Identity</b></p>
<p style = 'font-size:16px;font-family:Arial;color:#00233C'>i.e find the customers which were browing the site before and then logged-in. For this we can analyse the page table and visitor table.</p>

```
select sessionkey, count(distinct pagelocationdomain), count(distinct customerid) from demo_dim.page group by 1 order by 2 desc,1 desc ;
```

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>Let us check the page events and visitor table for one Sessionkey. <i>(We chose '98c93f0a-210d-48ec-8e11-2377598189f0' because of the activity prior to logging in.  It is the 2nd SessionKey above having 6 PageLocationDomain values)</i></p>

```
select * from demo_dim.page where sessionkey = '98c93f0a-210d-48ec-8e11-2377598189f0' order by eventtimestamp;
```

```
select * from demo_dim.visitor where sessionkey = '98c93f0a-210d-48ec-8e11-2377598189f0' order by eventtimestamp;
```

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>From the above output we can see that customer logged in the session and its CustomerId column is populated. We can check their browsing history before they logged-in in the page table.</p>

<p style = 'font-size:20px;font-family:Arial'><b>Customer tracking breaks from other organizations for cross domain identity</b></p>
<p style = 'font-size:16px;font-family:Arial;color:#00233C'> It is important to be able to recognize the customer even when they switch devices. In the sample data in the IndividualAttribute table, we have collected the AdobeMarketingCloudVisitor ID.</p>

```
select * from demo_dim.IndividualAttribute where sessionkey = '98c93f0a-210d-48ec-8e11-2377598189f0' order by eventtimestamp;
```

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>Above, we can see three different ID values in the AttributeValue column. This is not an issue for Celebrus as it creates new device id for unknown device and reassigns it customerid when he logs in. In the SystemIdentityAttributes table below, you'll see that a number id generated for an unknown device in the first row, and once the user logs in, the device is stitched to the customer ID enabling a full customer journey to be analyzed. </p>

```
select * from demo_dim.SystemIdentityAttributes where sessionkey = '98c93f0a-210d-48ec-8e11-2377598189f0' order by eventtimestamp;
```

<p style = 'font-size:20px;font-family:Arial'><b>Multi-Session Journey</b></p>

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>We can now use the Visitor table to track the customer across multiple sessions. As an example, we will use the CookieUniqueVisitorTrackingId from the Visitor table for customer CUST202210003 which occurred on 2 rows and then self join back to the visitor table to find all SessionKeys having the same  CookieUniqueVisitorTrackingId yielding 5 rows.  This would be part of a larger query to analyze what was being done across those sessions.</p>

```
select a.* from demo_dim.visitor a
join demo_dim.visitor b
on a.CookieUniqueVisitorTrackingId = b.CookieUniqueVisitorTrackingId and
b.customerid = 'CUST202210003' order by a.sessionkey,a.UVTPreviousSessionTimestamp;
```
