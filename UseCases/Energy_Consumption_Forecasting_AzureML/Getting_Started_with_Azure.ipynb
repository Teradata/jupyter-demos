<p style = 'font-size:16px;font-family:Arial;color:#00233C'>To get started with Azure and create an Azure account, follow these instructive steps:</p>
    <ol style = 'font-size:16px;font-family:Arial;color:#00233C'>
        <li>
            <b style = 'font-size:18px;font-family:Arial;color:#00233C'>Create an Azure Account:</b>
            <ul style = 'font-size:16px;font-family:Arial;color:#00233C'>
                <li>Go to <a href="https://azure.com" target="_blank">azure.com</a> in your web browser.</li>
                <li>Click on the "Start free" or "Create account" button (depending on the current Azure website design).</li>
                <li>You will be redirected to the Azure sign-up page.</li>
                <li>Sign in with your Microsoft account or create a new Microsoft account if you don't have one.</li>
                <li>Follow the on-screen instructions to provide the necessary information, including your email, phone number, and credit card details (required for account verification, but you won't be charged during the free trial period).</li>
            </ul>
        </li>
        <li>
            <b style = 'font-size:18px;font-family:Arial;color:#00233C'>Step-by-Step Guide:</b>
            <ul style = 'font-size:16px;font-family:Arial;color:#00233C'>
                <li>If you need more detailed instructions or assistance during the Azure account creation process, Microsoft provides a comprehensive step-by-step guide. You can access it by visiting <a href="https://learn.microsoft.com/en-us/training/modules/create-an-azure-account/" target="_blank">this link</a>.</li>
                <li>The guide will walk you through each stage, from sign-up to account verification, and provide additional information to ensure you successfully create your Azure account.</li>
            </ul>
        </li>
    </ol>

<hr>
<b style = 'font-size:18px;font-family:Arial;color:#00233C'>How to Create an Azure Workspace</b>
    <ol style="font-size: 16px; font-family: Arial;color:#00233C">
        <li>
            <strong>Navigate to the Azure Machine Learning Service:</strong>
            <ul>
                <li>Use the search bar to find "Machine Learning" or "Machine Learning Service" and click on the result to open the service's overview page.</li>
            </ul>
        </li>
        <li>
            <strong>Create a New Azure Machine Learning Workspace:</strong>
            <ul>
                <li>On the Azure Machine Learning Service overview page, click on the "Create" button to start creating a new workspace.</li>
                <img src = './images/fig1.png'>
            </ul>
        </li>
        <li>
            <strong>Fill in the Workspace Details:</strong>
            <p>Provide the required information for the new workspace:</p>
            <img src = './images/fig2.png'>
            <ul>
            <li><strong>Workspace Name:</strong> Choose a unique name for your workspace. This name must be globally unique across Azure.</li>
            <li><strong>Subscription:</strong> Select the Azure subscription you want to use for this workspace.</li>
            <li><strong>Resource Group:</strong> Either select an existing resource group or create a new one to organize resources related to this workspace.</li>
            <li><strong>Region:</strong> Choose the Azure region where you want to create the workspace. Select a region closest to your location for better performance.</li>
    </ul>
    </li>
        <li>
            <strong>Advanced Settings (Optional):</strong>
            <ul>
                <li>You can expand the "Advanced" tab to configure additional settings, if needed.</li>
            </ul>
        </li>
        <li>
            <strong>Review + Create:</strong>
            <ul>
                <li>Review your choices and click on the "Review + Create" button.</li>
            </ul>
        </li>
        <li>
            <strong>Validation and Creation:</strong>
            <ul>
                <li>Azure will validate your settings, and if everything is correct, click on the "Create" button to begin the workspace creation process.</li>
            </ul>
        </li>
        <li>
            <strong>Wait for Deployment:</strong>
            <ul>
                <li>Azure will start deploying your new Machine Learning workspace, and it might take a few minutes to complete.</li>
                <img src = './images/fig3.png'>
            </ul>
        </li>
        <li>
            <strong>Go to the Workspace:</strong>
            <ul>
                <li>Once the deployment is successful, click on the "Go to resource" button, or you can find your workspace on the Azure portal dashboard.</li>
                <img src = './images/fig4.png'>
            </ul>
        </li>
    </ol>
    <p style="font-size: 16px; font-family: Arial; color:#00233C">Congratulations! You have now created an Azure Machine Learning Workspace. This workspace will serve as the central hub for your machine learning projects, where you can develop, train, and deploy machine learning models using Azure services and tools. From here, you can access the Azure Machine Learning Studio, manage datasets, compute resources, and much more to build intelligent applications with ease.</p>

<hr>
<b style = 'font-size:18px;font-family:Arial;color:#00233C'>Access the Azure Machine Learning Studio:</b>
<img src = "./images/fig5.png">
<ol style="font-size: 16px; font-family: Arial;color:#00233C">
    <li>
        <strong>Create a Compute Instance:</strong>
    <ul>
        <li style="font-size: 16px; font-family: Arial;color:#00233C">In the Azure Machine Learning workspace overview page, click on the "Compute" option in the left-hand sidebar.</li>
        <li style="font-size: 16px; font-family: Arial;color:#00233C">On the "Compute Instances" page, click on the "New" button to create a new compute instance.</li>
        <img src = "./images/fig6.png">
        <li style="font-size: 16px; font-family: Arial;color:#00233C">Provide the following details for the new compute instance:</li>
        <img src = "./images/fig7.png">
            <ul style="font-size: 16px; font-family: Arial;color:#00233C">
                <li><strong>Compute Name:</strong> The compute name is the unique name given to the compute instance that you are creating. This name is used to identify the compute instance within your AzureML workspace.</li>
                <li><strong>Virtual Machine Type:</strong> You can choose between CPU and GPU capabilities.</li>
                <li><strong>Virtual Machine Size:</strong> Choose the size and compute power of the virtual machine for your instance. You can select from various machine types based on your needs.</li>
            </ul>
        <li style="font-size: 16px; font-family: Arial;color:#00233C">Click on the "Create" button to start creating the compute instance.</li>
    </ul>
    </li>
    <li>
        <strong>Wait for Deployment:</strong>
    <p style="font-size: 16px; font-family: Arial;color:#00233C">Azure will now start deploying your new compute instance. The process might take a few minutes to complete.</p>
    <h2 style="font-size: 16px; font-family: Arial;color:#00233C">Access the Compute Instance:</h2>
    <ul>
        <li style="font-size: 16px; font-family: Arial;color:#00233C">Once the deployment is successful, you can see the compute instance listed on the "Compute Instances" page.</li>
        <li style="font-size: 16px; font-family: Arial;color:#00233C">Click on the name of the compute instance to access its details.</li>
        <li style="font-size: 16px; font-family: Arial;color:#00233C">You can find the IP address and other information needed to connect to the compute instance.</li>
    </ul>
    </li>
    <p style="font-size: 16px; font-family: Arial;color:#00233C">That's it! You have now created a compute instance in AzureML, which you can use to run your data processing and machine learning tasks. The compute instance provides a pre-configured environment with popular machine learning libraries and tools, allowing you to focus on your data science work without worrying about infrastructure setup.</p>
