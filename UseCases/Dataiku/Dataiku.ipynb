<p style = 'font-size:20px;font-family:Arial;color:#E37C4D'><b>How to Setup Dataiku</b></p>
<p style = 'font-size:16px;font-family:Arial;color:#E37C4D'><b>Steps to establish connection to Teradata:</b></p>

<ol style = 'font-size:16px;font-family:Arial'>
  <li>Click on the "3x3" icon in the top right corner of the Dataiku home screen, then select "Administration" and "Connections".</li>
  <li>Click the "+ New Connection" button and select "Teradata" from the list of available connections.</li>
  <li>Enter a descriptive name for your connection in the "New connection name" field.</li>
  <li>In the "Host" field, enter the host address for your machine as shown on the ClearScape dashboard.</li>
  <li>Enter the name of your default database in the "Default database" field. This is typically "demo_user" for ClearScape users.</li>
  <li>Enter your username and password in the "User" and "Password" fields, respectively.</li>
  <li>In the "Advanced JDBC properties" field, enter the following properties:
    <ul>
      <li>CHARSET -> UTF8</li>
      <li>TMODE -> TERA</li>
    </ul>
  </li>
  <li>Tick the checkbox for "Autocommit mode" in the "Advanced params" section.</li>
  <li>Click the "TEST" button to test the connection. Make sure you are connected to your VPN before testing.</li>
  <li>If the test is successful, click the "CREATE" button to create the connection.</li>
</ol>
