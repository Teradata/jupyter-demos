```
create table ${workingDatabase}.${georefPrefixe}${coreTablePrefixe}${entityRef}
(geo_id              integer
,geo_name            varchar(200) character set unicode
,geo_ascii_name      varchar(200) character set latin
,geo_latitude        decimal(10,8)
,geo_longitude       decimal(11,8)
,geo_feature_code_id char(5)
,country_geo_id      integer
,admin_lv1_geo_id    integer
,admin_lv2_geo_id    integer
,population          bigint
,timezone_id         varchar(30)
)
unique primary index(geo_id)
;
```

```
insert into ${workingDatabase}.${georefPrefixe}${coreTablePrefixe}${entityRef}
select  cast(entity_ref.geonameid as integer)                                    as geo_id
       ,entity_ref.name                                                          as geo_name
       ,entity_ref.asciiname                                                     as geo_ascii_name
       ,cast(entity_ref.latitude                                                 as decimal(10,8)) as geo_latitude
       ,cast(entity_ref.longitude                                                as decimal(11,8)) as geo_longitude
       ,coalesce(entity_ref.feature_code,entity_ref.feature_class||'_ND','N_ND') as geo_feature_code_id
       ,cast(country_ref.geonameid as integer)                                   as country_geo_id
       ,cast(admin_lv1_ref.geonameid as integer)                                 as admin_lv1_geo_id
       ,cast(admin_lv2_ref.geonameid as integer)                                 as admin_lv2_geo_id
       ,cast(entity_ref.population as bigint)                                    as population
       ,entity_ref.timezone                                                      as timezone_id
from    ${workingDatabase}.${georefPrefixe}${stagingTablePrefixe}${entityRef} entity_ref
left outer join
        ${workingDatabase}.${georefPrefixe}${foreignTablePrefixe}${countryRef} country_ref
on      entity_ref.country_code = country_ref.iso
left outer join
        Unpack(on    ${workingDatabase}.${georefPrefixe}${foreignTablePrefixe}${adminLv1Ref}
               using TargetColumn   ('country_adminlv1_code')
                     OutputColumns  ('country_code'
                                    ,'admin_lv1_code'
                                    )
                     OutputDataTypes('varchar'
                                    ,'varchar'
                                    )
                     Delimiter      ('.')
                     Accumulate     ('geonameid')
              ) admin_lv1_ref
on      entity_ref.country_code = admin_lv1_ref.country_code
and     entity_ref.admin1_code  = admin_lv1_ref.admin_lv1_code
left outer join
        Unpack(on    ${workingDatabase}.${georefPrefixe}${foreignTablePrefixe}${adminLv2Ref}
               using TargetColumn   ('country_adminlv1_adminlv2_code')
                     OutputColumns  ('country_code'
                                    ,'admin_lv1_code'
                                    ,'admin_lv2_code'
                                    )
                     OutputDataTypes('varchar'
                                    ,'varchar'
                                    ,'varchar'
                                    )
                     Delimiter      ('.')
                     Accumulate     ('geonameid')
              ) admin_lv2_ref
on      entity_ref.country_code = admin_lv2_ref.country_code
and     entity_ref.admin1_code  = admin_lv2_ref.admin_lv1_code
and     entity_ref.admin2_code  = admin_lv2_ref.admin_lv2_code
;
```

```
create table ${workingDatabase}.${georefPrefixe}${coreTablePrefixe}${alternateNameRef}
(geo_id              integer
,geo_alternate_name  varchar(250) character set unicode
)
primary index(geo_id)
unique index(geo_id,geo_alternate_name)
;
```

```
insert into ${workingDatabase}.${georefPrefixe}${coreTablePrefixe}${alternateNameRef}
select  cast(geonameid as integer) as geo_id
       ,ngram                      as geo_alternate_name
from    NGramSplitter (on ${workingDatabase}.${georefPrefixe}${stagingTablePrefixe}${entityRef}
                       using  TextColumn          ('alternatenames')
                              Grams               ('1')
                              Delimiter           (',')
                              OutputTotalGramCount('true')
                              Accumulate          ('geonameid')
                      ) as alternate_name
;
```

```
create table ${workingDatabase}.${georefPrefixe}${coreTablePrefixe}${countryRef}
(geo_id               integer
,country_code_iso     char(2)
,country_code_iso3    char(3)
,country_code_iso_num smallint
,country_code_fips    char(2)
,capital_geo_id       integer
,continent_geo_id     integer
,area_sq_km           integer
)
unique primary index(geo_id)
unique index(country_code_iso)
unique index(country_code_iso3)
unique index(country_code_iso_num)
;
```

```
insert into ${workingDatabase}.${georefPrefixe}${coreTablePrefixe}${countryRef}
select  cast(country_ref.geonameid as integer)        as geo_id
       ,country_ref.iso                               as country_code_iso
       ,country_ref.iso3                              as country_code_iso3
       ,cast(country_ref."ISO-Numeric" as smallint)   as country_code_iso_num
       ,country_ref.fips                              as country_code_fips
       ,capital_entity_ref.geo_id                     as capital_geo_id
       ,continent_entity_ref.geo_id                   as continent_geo_id
       ,cast(country_ref."Area(in sq km)" as integer) as area_sq_km
from    ${workingDatabase}.${georefPrefixe}${foreignTablePrefixe}${countryRef} country_ref
left outer join
        ${workingDatabase}.${georefPrefixe}${coreTablePrefixe}${entityRef} capital_entity_ref
on      cast(country_ref.geonameid as integer) = capital_entity_ref.country_geo_id
and     capital_entity_ref.geo_feature_code_id = 'PPLC'
inner join
       (select  geo_id                           as geo_id
               ,case geo_name
                  when 'Europe'        then 'EU'
                  when 'Africa'        then 'AF'
                  when 'Asia'          then 'AS'
                  when 'North America' then 'NA'
                  when 'South America' then 'SA'
                  when 'Oceania'       then 'OC'
                  when 'Antarctica'    then 'AN'
                end                              as continent
        from    ${workingDatabase}.${georefPrefixe}${coreTablePrefixe}${entityRef}
        where   geo_feature_code_id = 'CONT'
       ) continent_entity_ref
on      country_ref.continent = continent_entity_ref.continent
;
```

```
-- volatile table for finding missing capital cities
create volatile table vol_missing_capital as
(select country_geo_id
       ,candidate_geo_id
from   (select  country_core_ref.geo_id        as country_geo_id
               ,country_ft_ref.country         as country_name
               ,country_ft_ref.capital         as country_capital_name
               ,entity_ref.geo_id              as candidate_geo_id
               ,entity_ref.geo_name            as candidate_geo_name
               ,entity_ref.geo_feature_code_id as candidate_geo_feature_code_id
               ,case
                 when candidate_geo_feature_code_id = 'ADM2' then 1 else 0
               end as feature_code_score
               ,entity_ref.population          as candidate_population
        from    ${workingDatabase}.${georefPrefixe}${coreTablePrefixe}${countryRef} country_core_ref
        inner join
                ${workingDatabase}.${georefPrefixe}${foreignTablePrefixe}${countryRef} country_ft_ref
        on      country_core_ref.geo_id = cast(country_ft_ref.geonameid as integer)
        inner join
                ${workingDatabase}.${georefPrefixe}${coreTablePrefixe}${alternateNameRef} alternate_name_ref
        on      country_ft_ref.capital = alternate_name_ref.geo_alternate_name
        inner join
                ${workingDatabase}.${georefPrefixe}${coreTablePrefixe}${entityRef} entity_ref
        on      country_core_ref.geo_id = entity_ref.country_geo_id
        and     alternate_name_ref.geo_id = entity_ref.geo_id
        where   country_core_ref.capital_geo_id is null

        union

        select  country_core_ref.geo_id        as country_geo_id
               ,country_ft_ref.country         as country_name
               ,country_ft_ref.capital         as country_capital_name
               ,entity_ref.geo_id              as candidate_geo_id
               ,entity_ref.geo_name            as candidate_geo_name
               ,entity_ref.geo_feature_code_id as candidate_geo_feature_code_id
               ,case
                  when candidate_geo_feature_code_id = 'ADM2' then 1 else 0
                end as feature_code_score
               ,entity_ref.population          as candidate_population
        from    ${workingDatabase}.${georefPrefixe}${coreTablePrefixe}${countryRef} country_core_ref
        inner join
                ${workingDatabase}.${georefPrefixe}${foreignTablePrefixe}${countryRef} country_ft_ref
        on      country_core_ref.geo_id = cast(country_ft_ref.geonameid as integer)
       inner join
                ${workingDatabase}.${georefPrefixe}${coreTablePrefixe}${entityRef} entity_ref
        on      country_core_ref.geo_id = entity_ref.country_geo_id
        and     country_ft_ref.capital = entity_ref.geo_name
        where   country_core_ref.capital_geo_id is null
        and     country_ft_ref.capital is not null
        ) x
        qualify row_number() over (partition by country_geo_id
                                   order by     feature_code_score desc
                                               ,candidate_population desc
                                  ) = 1 
)
with data
unique primary index (country_geo_id)
on commit preserve rows
;
```

```
update ${workingDatabase}.${georefPrefixe}${coreTablePrefixe}${countryRef}
from   vol_missing_capital
set    capital_geo_id = vol_missing_capital.candidate_geo_id
where  ${workingDatabase}.${georefPrefixe}${coreTablePrefixe}${countryRef}.geo_id         = vol_missing_capital.country_geo_id
;
```

```
create table ${workingDatabase}.${georefPrefixe}${coreTablePrefixe}${airportRef}
(geo_id               integer
,airport_icao_code    varchar(7)
,airport_iata_code    char(3)
,airport_municipality varchar(50) character set unicode
,airport_type         varchar(14)
)
primary index(geo_id)
unique index(airport_icao_code)
;
```

```
insert into ${workingDatabase}.${georefPrefixe}${coreTablePrefixe}${airportRef}
select  airport_candidate.geo_id        as geo_id
       ,airport_ref.icao_code           as airport_icao_code
       ,airport_ref.iata_code           as airport_iata_code
       ,airport_ref.municipality        as airport_municipality
       ,airport_ref.airport_type        as airport_type
from   (select  icao_code                                                              as icao_code
               ,iata_code                                                              as iata_code
               ,country_code_iso                                                       as country_code_iso
               ,airport_type                                                           as airport_type
               ,municipality                                                           as municipality
               ,cast(latitude  as decimal(10,8))                                       as latitude
               ,cast(longitude as decimal(11,8))                                       as longitude
               ,  cast((cast(latitude  as decimal(10,8))  +  90) /10 as integer) * 100
                + cast((cast(longitude as decimal(11,8)) + 180) /10 as integer)        as tesselation_id
        from    Unpack(on   (select  ident                        as icao_code
                                    ,nullif(iata_code,'0')        as iata_code
                                    ,iso_country                  as country_code_iso
                                    ,municipality                 as municipality
                                    ,"type"                       as airport_type
                                    ,oreplace(coordinates,'"','') as coordinates
                             from    ${workingDatabase}.${georefPrefixe}${foreignTablePrefixe}${airportRef}
                             where   "type" <> 'closed'
                            )
                       using TargetColumn   ('coordinates')
                             OutputColumns  ('longitude'
                                            ,'latitude'
                                            )
                             OutputDataTypes('varchar'
                                            ,'varchar'
                                            )
                             Delimiter      (',')
                             Accumulate     ('icao_code'
                                            ,'iata_code'
                                            ,'country_code_iso'
                                            ,'municipality'
                                            ,'airport_type'
                                            )
                      ) ar
       ) airport_ref
inner join
        ${workingDatabase}.${georefPrefixe}${coreTablePrefixe}${countryRef} country_ref
on      airport_ref.country_code_iso = country_ref.country_code_iso
inner join
       (select  geo_id                                            as geo_id
               ,geo_name                                          as geo_name
               ,country_geo_id                                    as country_geo_id
               ,geo_latitude                                      as geo_latitude
               ,geo_longitude                                     as geo_longitude
               , cast((geo_latitude  +  90) /10 as integer) * 100
                +cast((geo_longitude + 180) /10 as integer)       as tesselation_id
        from    ${workingDatabase}.${georefPrefixe}${coreTablePrefixe}${entityRef}
        where   geo_feature_code_id = 'AIRP'
       ) airport_candidate
on      country_ref.geo_id         = airport_candidate.country_geo_id
and     airport_ref.tesselation_id = airport_candidate.tesselation_id
where   SYSSPATIAL.SphericalDistance(airport_ref.latitude
                                     ,airport_ref.longitude
                                     ,airport_candidate.geo_latitude
                                     ,airport_candidate.geo_longitude
                                     ) / 1000 < 100
qualify row_number() over (partition by icao_code order by SYSSPATIAL.SphericalDistance(airport_ref.latitude
                                                                                       ,airport_ref.longitude
                                                                                       ,airport_candidate.geo_latitude
                                                                                       ,airport_candidate.geo_longitude
                                                                                       ) / 1000) = 1
;
```

```
-- identifiyng rows with duplicate iata_code
create volatile table vol_dup_iata as
(select  airport_icao_code
 from    ${workingDatabase}.${georefPrefixe}${coreTablePrefixe}${airportRef} ar
 where   airport_iata_code is not null
 qualify row_number () over (partition by airport_iata_code
                             order by     airport_icao_code
                            ) > 1
)
with data
unique primary index(airport_icao_code)
on commit preserve rows
;
```

```
-- delete duplicate iata_code for this value to be unique
update ${workingDatabase}.${georefPrefixe}${coreTablePrefixe}${airportRef}
from   vol_dup_iata
set    airport_iata_code = null
where  ${workingDatabase}.${georefPrefixe}${coreTablePrefixe}${airportRef}.airport_icao_code = vol_dup_iata.airport_icao_code
;
```
