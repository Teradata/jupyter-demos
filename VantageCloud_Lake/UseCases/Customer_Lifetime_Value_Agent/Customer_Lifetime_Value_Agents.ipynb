<p style="font-size:20px;font-family:Arial"><b>Introduction</b></p>

<p style="font-size:16px;font-family:Arial">
Customer Lifetime Value (CLV) measures the total net revenue a customer is expected to generate throughout their relationship with a business. It is a foundational metric for customer-centric organizations, guiding decisions in marketing, retention, service design, and product strategy. While traditional AI models focus primarily on predicting CLV, modern agentic AI systems go a step further by not only forecasting value but also proactively identifying actions and interventions that increase it.</p>
<p style="font-size:20px;font-family:Arial"><b>Why CLV Matters</b></p>
<p  style="font-size:16px;font-family:Arial">Most enterprises already maintain some form of CLV—but many struggle to translate it into operational insights or strategic decisions. As a result, the potential of CLV often remains underutilized. Agentic AI transforms this landscape by making CLV actionable, enabling teams to ask and answer complex business questions instantly. Examples include:
<ul style="font-size:16px;font-family:Arial">
    <li>Give me tables with customer information.</li>
    <li>What is the relationship between CLV and churn?</li>
    <li>Provide strategies to reduce churn.</li>
    </ul>
<p style="font-size:16px;font-family:Arial">    Through automated reasoning, contextual understanding, and tool-based execution, agentic systems empower business users to extract insights, explore scenarios, and drive meaningful outcomes without relying on manual analytics or specialist intervention.</p>

<p style = 'font-size:20px;font-family:Arial'><b>2. Architecture of this Customer Lifetime Value Agent</b></p>
<p style = 'font-size:16px;font-family:Arial'> Here's the flow of the Customer Lifetime Value Agent we are going to execute:<br>
    <img src="./images/clv_arch.jpg" alt="clv architecture" style="width: 120%; border: 4px solid #404040; border-radius: 10px;"/> 

<p style = 'font-size:16px;font-family:Arial'> The first node of the agentflow is the <b>Start</b> node. This node is where you set up variables that your agent will use throughout its workflow. Think of these variables as containers of information that the agent will need to access as it works through the flow. We have defined four variables: 
    <ol style = 'font-size:16px;font-family:Arial'>
    <li>main_agent_op -This is an output variable that will store the main response or result from the agent.  </li>
    <li>main_vis_op - This variable will hold output meant for visualization, like charts or graphs, which future agents can use. </li>
        <li> main_question - This variable stores the user’s main question or request.</li>
        <li>db_name - This holds the name of the database the agent will use. In this case, it is set to FINSERV.</li>
        </ol>
        </p>

<hr style="height:1px;border:none">
<p style = 'font-size:16px;font-family:Arial'> The second node of the agentflow is the condition node named <b>Check Question Intent</b>. A <b>Condition Agent</b> helps your flow make decisions based on what the user asks. It checks the user’s 
questions or input and decides which path the flow should follow next. Think of it as a traffic controller that 
directs questions to the right place. We have divided user intent for question in three scenarios below: 
     <ol style = 'font-size:16px;font-family:Arial'>
    <li>Scenario-0: User is asking for basic details about tables, databases, columns, etc.</li>
    <li>Scenario-1: User is asking about customer lifetime value, CLV, correlation, sentiments analysis, charts like pie chart, histogram etc. </li>
        <li>Scenario-2:  User is asking about features impacting churn,  recommendations to reduce the churn, what are the main factors leading to a Churn or low CLV, recommend me method to reduce Churn, which are important 
factors for Churn. </li>
        </ol>
        <p style = 'font-size:16px;font-family:Arial'>  Based on each scenario the execution of agents happen at runtime
        </p>

<hr style="height:1px;border:none">
<p style = 'font-size:16px;font-family:Arial'> Then there are three agents <b> Data Exploration Agent
    , Data Insights Agent and Stretgic Agent </b> attached to each scenario we have discussed in condition node. The content, flow and variable information is pre-filled in each of these agents. We have to configure some of the parameters which we will explain later in notebook. </p>

<hr style="height:1px;border:none">
<p style = 'font-size:16px;font-family:Arial'> After that we again have a condition agent named <b>Check if Chart Required </b> which checks if the response to the user require any visualization or not. Depending upon the output from this node the execution flow is directed towards <b>Visualization Agent</b> or towards <b>Generate the Response
    </b> node.<br> The Visualization Agent will create a json required for the chart type required from the output for the Draw Plot function to display as reponse for the user query.<br>The last node in the flow is <b > Generate the Response </b> which is a LLM node. The LLM Node uses a language model to generate smart responses based on the user's 
input and the instructions you provide. </p>
