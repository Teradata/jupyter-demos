<p style="font-size:20px;font-family:Arial"><b>Vector Search Algorithms</b></p>

<p style="font-size:16px;font-family:Arial">
Teradata supports three main search algorithms:
</p>

<p style="font-size:18px;font-family:Arial"><b>1. VectorDistance</b></p>
<ul style="font-size:16px;font-family:Arial">
  <li><b>Flat index</b> (performs a full-table scan)</li>
  <li><b>Scalable and parallel search</b></li>
  <li><b>Exact search approach</b></li>
  <li>Suitable for <b>smaller datasets (<10K documents)</b></li>
</ul>

<p style="font-size:18px;font-family:Arial"><b>2. Kmeans</b></p>
<ul style="font-size:16px;font-family:Arial">
  <li><b>IVF (Inverted Vector File) index</b> (organizes data into clusters)</li>
  <li><b>Scalable and parallel training/search</b></li>
  <li><b>Approximate search approach</b></li>
  <li>Suitable for <b>larger datasets (>10K documents)</b></li>
</ul>

<p style="font-size:18px;font-family:Arial"><b>3. HNSW</b></p>
<ul style="font-size:16px;font-family:Arial">
  <li><b>HNSW (Hierarchical Navigable Small World) index</b></li>
  <li><b>Graph-based Approximate Nearest Neighbors (ANN) search</b></li>
  <li><b>Scalable and parallel training/search</b></li>
  <li><b>Approximate search approach</b></li>
  <li>Suitable for <b>larger datasets (>10K documents)</b></li>
</ul>

<p style="font-size:18px;font-family:Arial"><b>Checking the Status of the vector store</b>
<p style="font-size:16px;font-family:Arial">    
Though the create, update, destroy APIs are synchronous and will complete only after the operation is complete, user can ensure the same using the <b>status</b> API.  <br>
    Check the <b>status</b>: The status function allows us to confirm whether the particular operation is successful.</p>

```python
vs.status()
```

<p style="font-size:16px;font-family:Arial"><b> Wait for the status to become Ready to use the vector store.</b><br>We create the below function which will wait till the status of the Vector Store is in Ready State.
</p>

```python
def check_vs_ready(vs, poll_interval=15, timeout=600):
    """
    Waits until vector store status becomes READY.
    Args:
        vs: object having .status() method returning DataFrame
        poll_interval (int): seconds between checks
        timeout (int): max wait time in seconds
    Returns:
        DataFrame row when READY
    Raises:
        TimeoutError if not ready within timeout
    """
    start_time = time.time()

    while True:
        df = vs.status()
        if df is not None and not df.empty:
            status = df.loc[0, 'status']
            if status == 'READY':
                print(f"The Vector Store Database: {df.loc[0,'vs_name']} is READY!")
                return df
            else:
                print(f"Current status: {status}. Waiting {poll_interval} seconds...")
        else:
            print("Status not available yet. Retrying...")
        # Timeout check
        if time.time() - start_time > timeout:
            raise TimeoutError("Vector Store did not become READY within timeout")
        time.sleep(poll_interval)
```

```python
check_vs_ready(vs)
```

<p style="font-size:18px;font-family:Arial"><b>6.2 Performing a Similarity Search in a Vector Store</b>
<p style="font-size:16px;font-family:Arial"> <b> similarity search</b> allows us to find the most relevant vectors based on a given input. By passing a question to an existing vector store, we can retrieve the closest matches based on the selected search algorithm.<br>
<ol style="font-size:16px;font-family:Arial"><b>How It Works</b>
    <li><b>Input a question</b> Pass a question or input text to question argument of similarity_search().</li>
        <li><b>Perform the Similarity Search</b> The vector store searches for the most relevant matches using the specified search_algorithm.</li>
        <li><b>Retrieve Top Matches</b> The result contains the top_k most relevant entries along with their similarity scores, helping us understand how closely they match the input.</li>
        </ol>
 <p style="font-size:16px;font-family:Arial">   This process enables efficient retrieval of information based on vector embeddings, making it useful for applications such as <b>semantic search, recommendation systems, and document retrieval</b>.     

```python
question = 'Which item has most comments'
response = vs.similarity_search(question=question)
```

```python
# The "similar_objects" in the response is by default a DataFrame with top_k entries.
response.similar_objects
```

```python
# Alternative option: Using JSON format could be speedier.
# We repeat the previous call by now specifying return_type='json'
question = 'Which item has most comments?'
response = vs.similarity_search(question=question, return_type='json')
```

```python
# The "similar.objects" in the response is now a JSON object with top_k entries.
response.similar_objects
```

<p style="font-size:20px;font-family:Arial"><b>9. Metadata based Vector Store </b></p>
<p style="font-size:16px;font-family:Arial">Metadata-based vector store contains metadata information about tables such as the table and column description. The description is populated using an LLM via prompt and optionally samples from table contents and other tables in the vector store are used to generate it.
</p> 

<hr style="height:1px;border:none;">
<p style="font-size:18px;font-family:Arial"><b>9.1  Creating Metadata based Vector Store </b>
<p style="font-size:16px;font-family:Arial">Metadata-based vector store can be formed in two ways:<br>
    <ol style="font-size:16px;font-family:Arial">
        <li>Specify the table names in include_objects parameter which the user wants to include to create Vector Store and if the user wants to exclude specific set of tables they can specify the same in exclude_objects parameter.</li>
        <li>Specify the patterns in include_patterns and exclude_patterns parameter.</li>
</ol>

<p style="font-size:18px;font-family:Arial"><b>9.1.1  Creating Metadata based Vector Store using tablenames</b>
