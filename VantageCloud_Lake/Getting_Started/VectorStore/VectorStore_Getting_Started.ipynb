<p style="font-size:20px;font-family:Arial"><b>Introduction</b></p>

<p style="font-size:16px;font-family:Arial">
Vector stores are specialized databases designed for efficient storage, indexing, and searching of high-dimensional vector embeddings. These embeddings, generated by AI models, enable similarity search and power applications in machine learning, NLP, recommendation systems, and image/video retrieval.

<p style="font-size:16px;font-family:Arial">

<hr>

<p style="font-size:20px;font-family:Arial"><b>Vector Search Algorithms</b></p>

<p style="font-size:16px;font-family:Arial">
Teradata supports three main search algorithms:
</p>

<p style="font-size:18px;font-family:Arial"><b>1. VectorDistance</b></p>
<ul style="font-size:16px;font-family:Arial">
  <li><b>Flat index</b> (performs a full-table scan)</li>
  <li><b>Scalable and parallel search</b></li>
  <li><b>Exact search approach</b></li>
  <li>Suitable for <b>smaller datasets (<10K documents)</b></li>
</ul>

<p style="font-size:18px;font-family:Arial"><b>2. Kmeans</b></p>
<ul style="font-size:16px;font-family:Arial">
  <li><b>IVF (Inverted Vector File) index</b> (organizes data into clusters)</li>
  <li><b>Scalable and parallel training/search</b></li>
  <li><b>Approximate search approach</b></li>
  <li>Suitable for <b>larger datasets (>10K documents)</b></li>
</ul>

<p style="font-size:18px;font-family:Arial"><b>3. HNSW</b></p>
<ul style="font-size:16px;font-family:Arial">
  <li><b>HNSW (Hierarchical Navigable Small World) index</b></li>
  <li><b>Graph-based Approximate Nearest Neighbors (ANN) search</b></li>
  <li><b>Scalable and parallel training/search</b></li>
  <li><b>Approximate search approach</b></li>
  <li>Suitable for <b>larger datasets (>10K documents)</b></li>
</ul>

<p style="font-size:16px;font-family:Arial">Steps to create Vector Store

```python
# ACTION: Specify a name for a new vector store.
vs = VectorStore(name="vs_comment", log=True)
```

```python
# We need to specify the data source to create the embeddings for the 
# new vector store. The data source can be a string of a (fully-qualified, 
# if needed) table name, or a teradataml DataFrame.
#
# The following creates a teradataml DataFrame from this table to use as input.
# As the usecase is only for knowledge purpose we are taking subset of data
input_data = DataFrame.from_query("SELECT * FROM DEMO_Retail.Web_Comment where comment_id < 1000")
input_data
```

```python
# Create vector store with the title specified earlier.
# The following specification lacks "chat_completion_model" argument.
# Per the user guide, default value is "anthropic.claude-3-haiku-20240307-v1:0"
# NOTE: If you ask to create an existing vector store, an error will be thrown.
vs.create(embeddings_model= 'amazon.titan-embed-text-v1',
          search_algorithm= 'KMEANS',
          seed=10,
          top_k=5,
          metric='EUCLIDEAN',
          object_names= input_data,
          key_columns= ['comment_id'],
          data_columns= ['comment_text'],
          vector_column= 'VectorIndex')
```

<p style="font-size:18px;font-family:Arial"><b>5.1 Performing a Similarity Search in a Vector Store</b>
<p style="font-size:16px;font-family:Arial"> <b> similarity search</b> allows us to find the most relevant vectors based on a given input. By passing a question to an existing vector store, we can retrieve the closest matches based on the selected search algorithm.<br>
<ol style="font-size:16px;font-family:Arial"><b>How It Works</b>
    <li><b>Input a question</b> Pass a question or input text to question argument of similarity_search().</li>
        <li><b>Perform the Similarity Search</b> The vector store searches for the most relevant matches using the specified search_algorithm.</li>
        <li><b>Retrieve Top Matches</b> The result contains the top_k most relevant entries along with their similarity scores, helping us understand how closely they match the input.</li>
        </ol>
 <p style="font-size:16px;font-family:Arial">   This process enables efficient retrieval of information based on vector embeddings, making it useful for applications such as <b>semantic search, recommendation systems, and document retrieval</b>.     

```python
question = 'Which item has most comments'
response = vs.similarity_search(question=question)
```

```python
# The "similar_objects" in the response is by default a DataFrame with top_k entries.
response.similar_objects
```

```python
# Alternative option: Using JSON format could be speedier.
# We repeat the previous call by now specifying return_type='json'
question = 'Which item has most comments?'
response = vs.similarity_search(question=question, return_type='json')
```

```python
# The "similar.objects" in the response is now a JSON object with top_k entries.
response.similar_objects
```

<hr style="height:1px;border:none;">
<p style="font-size:18px;font-family:Arial"><b>5.2 Generating a Natural Language Response</b>
<p style="font-size:16px;font-family:Arial"> Once the <b> similarity search</b>method retrieves relevant results from the vector store, we can generate a natural language response by proving a prompt.<br>
<ol style="font-size:16px;font-family:Arial"><b>Steps to Prepare the Response</b>
    <li><b>Input a question</b> Pass a question or input text to question argument of prepare_response(). </li>
    <li><b>Retrieve Similarity Results</b> The similarity_search() method returns the most relevant matches along with their similarity scores.</li>
        <li><b>Input a prompt</b> By passing prompt, we ask the model to format the output in a particular way.</li>
        </ol>
 <p style="font-size:16px;font-family:Arial">   As we can see below, by passing the question and the prompt, the response is formated in a conversational manner as suggested in the prompt.</p>.     

```python
question='Did any one say the about the material of shirt?'
prompt='Format the response in a conversational way.'
response = vs.prepare_response(question=question, similarity_results=response, prompt=prompt)
```

```python
# NOTE: The "response" object is a string. If we ask to display the string itself,
# the special characters like new lines will not be interpreted.
# To show the actual text with new lines, explicitly specify to print() the "response" object.

print(response)
```

<hr style="height:1px;border:none;">
<p style="font-size:18px;font-family:Arial"><b>5.3 ask -Unified Similarity Search & Response Generation</b>
<p style="font-size:16px;font-family:Arial"> To improve efficiency, we can  <b> combine similarity search and response generation</b> into a single operation. This method performs a similarity search in the vector store using the <b>input query</b> and then prepares a natural language response based on the retrieved results.<br>
<ol style="font-size:16px;font-family:Arial"><b>How It Works</b>
    <li><b>Input a question</b> Pass a question or input text to question argument of ask(). </li>
    <li><b>Input a prompt</b> By passing prompt, we ask the model to format the output in a particular way.</li>
        </ol>
 <p style="font-size:16px;font-family:Arial"> <b>Faster Execution</b>  By combining both steps into a single function call, response time is reduced, making the system more efficient.</p>  

```python
custom_prompt = """Do not assume information. 
Only provide information that is present in the data.
Format results like this:
Comment ID: 
Comment: 
"""
question ='Are there any negative comments on shirts?'
response = vs.ask(question=question, prompt=custom_prompt)
```

```python
print(response)
```
