## Setting up the data

### Load Data and DataFrame Objects

In this example we will load some tables using teradataml. As we have already logged in, we can call teradataml to read a CSV file and create a SQL table:

```python
from teradataml import *
import pandas as pd

df = pd.read_csv("ADS_Final_Test.csv")
copy_to_sql(df, "ADS_Final_Test")

# Open python script and update based on default database name
with open("script_example.py") as f:
    newText=f.read().replace('<db_name>', default_database)
with open("ex1pScoViaDSS.py", "w") as f:
    f.write(newText)
```

## Example: Uploading Files

Run the cell and in the UI do the following:

1. Go to Files 
2. Set ID to <b>ex1pScoViaDSS</b> and Name to <b>ex1pScoViaDSS.py</b> and turn off Binary, Turn on Replace and Click Install
3. Set ID to <b>ex1pModForDSS</b> and Name to <b>ex1pModForDSS.out</b> and turn on Binary, Turn on Replace and Click Install

```python
from teradatamlwidgets import script

ui = script.Ui(
    files_local_path='.', 
    script_command='echo upload')
```

## Execute

Run the cell and in the UI do the following:

1. Type in Script Command: <b>tdpython3  ./'+default_database+'/ex1pScoViaDSS.py</b>
2. Go to Output Tab and enter:
 * <b>Cust_ID</b> and <b>INTEGER</b> and Add
 * <b>Prob_0</b> and <b>FLOAT</b> and Add
 * <b>Prob_1</b> and <b>FLOAT</b> and Add
 * <b>Actual_Value</b> and <b>INTEGER</b> and Add       
3. Press Execute

```python
from teradatasqlalchemy import (CHAR, VARCHAR, CLOB, INTEGER, FLOAT)
from teradataml import *

ui = script.Ui(
    data=DataFrame('ADS_Final_Test'),
    script_name='execute',
    files_local_path='.', 
    script_command='tdpython3  ./'+default_database+'/ex1pScoViaDSS.py',
    search_path="myDB",
    returns={"Cust_ID": INTEGER(), "Prob_0": FLOAT(), "Prob_1": FLOAT(), "Actual_Value": INTEGER()})
```
