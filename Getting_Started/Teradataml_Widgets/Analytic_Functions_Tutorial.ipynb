## SQLE Example : Moving Average

For the function “MovingAverage”, set the parameters for the function to the following in this example:
1. In the **_Required_** tab_, click the **input** dropdown. 
- **Table** is “ibm_stock”<br>
- **Data Partition Option** will not be used because Data By and Order By have no values<br>


2. In the **_Optional_** tab
- **TargetColumns** needs to be set to “stockprice”. Click in the box and select it from the list.<br>
    - **IncludeFirst** should be selected.<br>
- **Alpha** should be set to 0.22<br>
- **StartRows** should be 3<br> 
- **WindowSize** should be set to 11<br>
- **MavgType** should be “M”

3. Click the **Execute** button.

```python
from teradatamlwidgets import analytic_functions

inputs = ['ibm_stock']
outputs = ['Project_OutMovingAverageTest']
ui = analytic_functions.Ui(function= 'MovingAverage',
        outputs=outputs, 
        inputs=inputs)
```

<hr style="height:2px;border:none;background-color:#00233C;">

## SQLE Example : Moving Average with parameters from a file

As explained in OPTIONAL: Load and Save section above, if the export_settings is set to a json, it will automatically load the parameters that are set in that JSON. You can then modify the parameters if needed and click Save button which will update the JSON accordingly.

In this example notice there is "export_settings" added in the constructor of analytic_functions.UI, this way you simply have to run the cell and the parameters will already be set.

The parameters this time are already set to the following below:
1. In the **_Required**_ tab<br>
- **Table** is set to “ibm_stock”. You will need to click the **input** dropdown to verify.<br>
- **Data Partition Option** is set to "None"

2. In the **_Optional_** tab<br>
- **TargetColumns** is set to “stockprice”
- **IncludeFirst** should be _selected_<br>
- **Alpha** is 0.22<br>
- **StartRows** is 3<br>
- **WindowSize** is 11<br>
- **Mavg Type** is “M”<br>

3. Click the **Execute** button.

```python
from teradatamlwidgets import analytic_functions

inputs = ['ibm_stock']
outputs = ['Project_OutMovingAverageTest']
ui = analytic_functions.Ui(function= 'MovingAverage',
        outputs=outputs, 
        inputs=inputs,  
        export_settings="data/MovingAverage.json")
```

## Pipeline Structure

## VAL Example: Linear Regression

The VAL examples of Linear Regression and then Linear Regression Predict demonstrates the possible pipeline structure. In other words we will create Linear Regression VAL output that will be used as an input to Linear Regression VAL Predict.

The code below has the input, output, function name, and file name specified already.
The filename will automatically populate the function with the parameters.

After running the cell, the parameters as mentioned before are already filled out. 

1. In the **_Required_** tab<br>
- **Table** is “titanic”. Click the **input_A_role** dropdown to verify.<br>
- **InputColumns** are “age” and “p_class”<br>
- **Response Column** is “fare”


2. In the **_Optional_** tab<br>
- There are no **Group Bycolumns**<br>
    - No check boxes are selected for the parameters<br>
- **Condition Index Threshold** is set to 37<br>
- **Entrance Criterion** is 3.84<br>
- **Remove Criterion** is 3.84<br>
- **Variance Proportion Threshold** is 0.5

3. Click the **Execute** button.

```python
from teradatamlwidgets import analytic_functions

inputs = ['titanic']
outputs = ['Project_OutLinearRegression']
ui = analytic_functions.Ui(outputs=outputs,  
        function="Linear Regression", 
        inputs=inputs,  
        export_settings="data/LinReg.json")
```

After you click Execute, you see the resulting table. We want to use this table as the input to the Linear Regression Predict in VAL.

## VAL Example: Linear Regression Predict

This is step 2 of the pipeline structure.
 
Notice that the input is Titanic but ALSO has “dssDB. Project_OutLinearRegression”, this was the output table name we specified in the previous cell for Linear Regression VAL. 

Once again, the export_settings has the saved values for the different parameters.

1. In the **_Required_** tab<br>
- The **input_A_role** dropdown has **Table** set to “titanic”<br>
- The **linear_regression** dropdown has **Table** set to “Project_OutLinearRegression”, which is the output from the previous cell.<br>

2. In the **_Optional_** tab<br>
- **Index Columns** has “SURVIVED” already selected<br>
- **Response Column** has “fare” already selected<br>
- **Accumulate** has “age” and “p_class” already selected<br>

3. Click the **Execute** button.

```python
from teradatamlwidgets import analytic_functions

inputs = ['titanic', 'Project_OutLinearRegression']
outputs = ['Project_OutLinearRegressionPredict']
ui2 = analytic_functions.Ui(outputs=outputs,
        function="Linear Regression Predict", 
        inputs=inputs,  
        export_settings="data/LinRegPredict.json")
```

After executing, this is the Linear Regression Prediction. This was calculated using the Linear Regression VAL function we created above.
