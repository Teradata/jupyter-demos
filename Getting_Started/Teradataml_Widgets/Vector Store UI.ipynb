### Example of Vector Store use

This example will work through the whole UI as an example for how to use the UI.

This assumes use of an AWS connection.

#### For Creation:
<u>Step 1</u>:

- Click the Health heart button to see health status of the Vector Store

  
<u>Step 2</u>:

- Click the Add button on left panel to create new vector table

    In the Names section:
    
        - Set “Vector Store Name” to any name, for instance “vector_store_example”
        
        - Set “Object Names” to "oaf"."amazon_reviews_25"
        
        - In the Columns section:
        
        - Set “Data Columns” to rev_text

<br>

- Click the green Create button

- Click the information icon to the right of the Create button to see the status of the vector table, which should show as READY

<u>Step 3</u>:

- Click the Refresh button to see the new vector table in list


<u>Step 4</u>:

- Click the newly created vector table and click the Use check button


#### To Use and Ask questions:

<b>Example 1</b>:

- For the Question enter “Which book are all the reviews talking about?”

- Click check box for Similarity Search 

- Click green Run button


<b>Example 2</b>:

- For the Question enter 'Did any one feel this is a thin book?'

- For the Prompt enter 'Format the response in a conversational way.'

- Click check box for Prepare Response (both Similarity Search and Prepare Response should be selected)

- Click green Run button


<u>Step 5</u>: 

- Click the Details button to see details of vector table

<u>Step 6</u>: 

- Click Update button

      - Add Description as anything, for instance “This is an example for teradatamlwidgets Vector Store”
<br>
- Click the green Update button
