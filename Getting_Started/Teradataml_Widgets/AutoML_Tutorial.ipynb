<p style = 'font-size:20px;font-family:Arial'><b>Setting up the test train data.</b></p>
<p style = 'font-size:18px;font-family:Arial'><b>Load Data and DataFrame Objects.</b></p>
<p style = 'font-size:16px;font-family:Arial'>In this example we will load some tables using teradataml. As we have already logged in, we can call teradataml load functions.>/p>

```python
from teradataml import load_example_data
from teradataml import DataFrame
load_example_data("teradataml", "iris_input")
iris = DataFrame.from_table("iris_input")
iris.head()
```

<p style = 'font-size:18px;font-family:Arial'><b>Test Train Split</b></p>

```python
iris_sample = iris.sample(frac = [0.8, 0.2])
# Fetching train and test data
iris_train= iris_sample[iris_sample['sampleid'] == 1].drop('sampleid', axis=1)
iris_test = iris_sample[iris_sample['sampleid'] == 2].drop('sampleid', axis=1)
from teradataml import copy_to_sql
copy_to_sql(df = iris_train, table_name = "iris_train", if_exists="replace")
copy_to_sql(df = iris_test, table_name = "iris_test", if_exists="replace")
```

<p style = 'font-size:20px;font-family:Arial'><b>Example: AutoML Classification Standalone</b></p>
<p style = 'font-size:18px;font-family:Arial'>Run the cell to open the AutoML UI. Then follow these steps:</p>

<ol style = "font-size:16px;font-family:Arial">
    <li>In Initialize tab, set Target Column to species and click Fit</li>
    <li>After Fit has completed, click the Close button.</li>
    <li>Then click Leaderboard to see the top leaders. Then click the Close button.</li>
    <li>In the Prediction tab, click Execute to calculate the predicted output. Clicking the Close button will return you to the AutoML configuration options.</li>
</ol>    

```python
from teradatamlwidgets import auto_ml

ui = auto_ml.Ui(
    task="Classification", 
    training_table=iris_train, 
    predict_table='iris_input',
    algorithms=['xgboost', 'knn'],
    verbose=0,
    max_runtime_secs=300,
    max_models=5)
```
