## Setting up the test train data

### Load Data and DataFrame Objects

In this example we will load some tables using teradataml. As we have already logged in, we can call teradataml load functions:

```python
from teradataml import load_example_data
from teradataml import DataFrame
load_example_data("teradataml", "iris_input")
iris = DataFrame.from_table("iris_input")
iris.head()
```

### Test Train Split

```python
iris_sample = iris.sample(frac = [0.8, 0.2])
# Fetching train and test data
iris_train= iris_sample[iris_sample['sampleid'] == 1].drop('sampleid', axis=1)
iris_test = iris_sample[iris_sample['sampleid'] == 2].drop('sampleid', axis=1)
from teradataml import copy_to_sql
copy_to_sql(df = iris_train, table_name = "iris_train", if_exists="replace")
copy_to_sql(df = iris_test, table_name = "iris_test", if_exists="replace")
```

## Example: AutoML Classification Standalone

Run the cell to open the AutoML UI. Then follow these steps:

1. In Initialize tab, set Target Column to species and click Fit
2. After Fit has completed, click the Close button.
3. Then click Leaderboard to see the top leaders. Then click the Close button.
4. In the Prediction tab, click Execute to calculate the predicted output. Clicking the Close button will return you to the AutoML configuration options.

```python
from teradatamlwidgets import auto_ml

ui = auto_ml.Ui(
    task="Classification", 
    training_table=iris_train, 
    predict_table='iris_input',
    algorithms=['xgboost', 'knn'],
    verbose=0,
    max_runtime_secs=300,
    max_models=5)
```
