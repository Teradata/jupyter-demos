## Example: DataRobotPredict

Score data in Vantage with a model that has been created outside the Vantage by removing all the all cached models.

The parameters for the DataRobotPredict example are as follows:

- Set Accumulate to 'id', 'sepal_length', 'petal_length'
- Set OverwriteCachedModel to "*"
  
- Function is already set to DataRobotPredict
- BYOM Install is already set to "mldb"
- InputTable is already set to "iris_test"
- ModelID is already set to "dr_iris_rf"
- ModelTable is already set to "byom_models"

```python
from teradatamlwidgets import byom_functions

ui = byom_functions.Ui(function = "DataRobotPredict", 
        byom_location = "mldb", 
        data="iris_test", 
        model_id="dr_iris_rf", 
        model_table="byom_models")
```
