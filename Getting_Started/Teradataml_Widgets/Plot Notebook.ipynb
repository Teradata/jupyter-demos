### Plot Notebook Tabs

The parameters in the tabs are from the TD_PLOT function. Please refer to that documentation for the specifics of each parameter (https://teraworks.teradata.com/pages/viewpage.action?spaceKey=TDANEX&title=TD_PLOT+Parameters#TD_PLOTParameters-MARKER)

#### Chart:
This tab includes the X value of the input table and the Y values(s) which are the Series[0], Series[1] etc. 
- Each time a Series is added, a new series option appears
- A tab is available for each series to customize

#### Format
This tab includes all the formatting specific parameters for the plot.
- Title, Heading, X-Label, Y-Label, and grid, color map and legend information

#### Range

Includes all the range specific parameters
- X-Ticks, Reverse X-Ticks, X Range (Min and Max)
- Y-Ticks, Reverse Y-Ticks, Y Range (Min and Max)

#### Series Tab(s)
Each Series tab has same parameters which is specific to that series.
- Style color, Line Style, Line Width, Marker, Marker Size

## Class: plot.Ui
## Constructor:

Description: Opens the Plot UI dialog in the notebook with the following arguments:

| Argument | Type | Required | Description | Example |
|:--------:|:--------:|:--------:|:--------:|:--------:|
|  table_name |  String   |  Mandatory (IF df argument not specified)  | An input table name to use for plotting. | Ui(table_name = "titanic") |
|  df |  teradataml.DataFrame   |  Mandatory (IF table_name argument not specified) | An input teradataml dataframe to use for plotting. | Ui(df = DataFrame("titanic")) |
|  current_plot   |  String   |  Optional   | If you want chart type pre-selected. | Ui(table_name = "titanic", current_plot = "Bar")|
|  _all other parameter arguments_   |     |  Optional   | If you want any other parameters pre-selected, see their argument name in description. | Ui(table_name = "titanic", current_plot = "Bar", style='green')|

### Setting up the Plot Notebook : Optional Arguments

Optionally, you can set up any of the parameters in the constructor itself. This way when running the cell, the UI gets populated with the pre-selected choices. Each parameter has an argument associated which is what would be used to set it in the constructor.

You can see the argument for a given parameter by hovering over the parameter, this will give description of the parameter and at the end specifies the argument name.

For example, below we see the Style Color parameter has argument: style, so in the constructor to set this to ‘green’ there is : 
style = ‘green’ 

This allows the UI to have that selection upon loading (i.e. when the notebook cell is run). 

The examples below will use this in order to get the parameters to be filled with the appropriate values.

#### Example 1 - Simple Line Plot

For the first plot below, we will do an example of a "Line" plot. This uses "ibm_stock" as the input table.

- In the _Chart Tab_, choose **X** as "period" and **Series[0]** as "stockprice". 

- In the _Format Tab_, for the **Title** enter "Line Plot Example" and for the **Heading** enter "Simple Line Plot".  

Once you are done entering the parameters, click the green Plot button.

```python
from teradatamlwidgets import plot

plot1 = plot.Ui(
    table_name="ibm_stock", 
    current_plot="Line", 
    x='period', 
    series='stockprice', 
    color='green', 
    heading='Simple Line Plot',
    title= 'Line Plot Example')
```

<hr style="height:2px;border:none;background-color:#00233C;">

#### Example 2 - Simple Bar Plot

For the second plot below, we will do an example of a "Bar" plot. This uses "ibm_stock" as the input table.

- In the _Chart Tab_, choose **X** as "period" and **Series[0]** as "stockprice". 

- In the _Format Tab_, for the **X Label** enter "xlabel" and for the **Y Label** enter "ylabel", for Heading enter “Simple Bar Plot”, and for Title enter “Bar Plot Example”

- In the _Range Tab_, for the **X Ticks** enter "MMM" and for the **Y Ticks** enter "9,99.99".

- In the _Series[0] Tab_, keep **Style Color** as "blue".

Once you are done entering the parameters, click the green Plot button.

```python
from teradatamlwidgets import plot

plot2 = plot.Ui(
    table_name="ibm_stock", 
    current_plot="Bar", 
    x='period', 
    series='stockprice', 
    xlabel='xlabel',
    ylabel='ylabel',
    x_ticks='MMM',
    y_ticks="9,99.99",
    color='blue', 
    heading='Simple Bar Plot',
    title= 'Bar Plot Example')
```


<hr style="height:2px;border:none;background-color:#00233C;">

#### Example 3 - Composite Bar Plot

The plot below is an example of a Composite Bar chart. It uses "us_air_pass" as the input table.

- In the _Chart Tab_, for the **X** choose "idx" and for **Series[0]** choose "international", and **Series[1]** set to "domestic".

- In the _Format Tab_, for **Heading** enter "Composite Bar Plot", and Title to “Bar Plot Example 2”.

- In _Series[0] Tab_ the **Style Color** is kept to the default of blue. But for _Series[1] Tab_ change **Style Color** to "orange".

Once you are done entering the parameters, click the green Plot button.

```python
from teradatamlwidgets import plot

plot3 = plot.Ui(
    table_name="us_air_pass", 
    current_plot="Bar", 
    x='idx', 
    series=['international','domestic'], 
    color=['blue', 'orange'], 
    heading='Composite Bar Plot',
    title= 'Bar Plot Example 2')
```

<hr style="height:2px;border:none;background-color:#00233C;">

#### Example 4 - Scatter Plot

The plot below is an example of a Scatter plot. It uses "iris_input" as the input table.

- In the _Chart Tab_, for the **X** choose "sepal_length" and for **Series[0]** choose "petal_length".

- In the _Format Tab_, for **Title** enter "Scatter Plot of sepal_length vs petal_length", set **Heading** to “Scatter Plot Example”, **X Label** to "sepal_length" and the **Y Label** to “petal_length”, set **Grid Color** to “black”, and set **Grid Width** to 1.

- In _Series[0] Tab_, keep **Style Color** as "blue".

Once you are done entering the parameters, click the green Plot button.

```python
from teradatamlwidgets import plot

plot4 = plot.Ui(
    table_name="iris_input", 
    current_plot="Scatter", 
    x='sepal_length', 
    series='petal_length', 
    xlabel='sepal_length',
    ylabel='petal_length',
    grid_color='black',
    grid_linewidth=1, 
    grid_linestyle="-",
    color='blue', 
    title='Scatter Plot of sepal_length vs petal_length',
    heading= 'Scatter Plot Example')
```

<hr style="height:2px;border:none;background-color:#00233C;">

#### Example 5 - Mesh Plot
Below is an example of a Mesh Plot, we use "waveletTable" as the input table.

- In the _Chart Tab_, set **X** to "x", **Scale** to "c", and **Series[0]** to "y".
- In the _Format Tab_, set **Title** to “Mesh Plot Example”, set **Heading** to “Simple Mesh Plot”, set **Color Map** to "ice", and set the **Min** to -0.5 and **Max** to 0.5.

```python
from teradatamlwidgets import plot

plot5 = plot.Ui(
    table_name="waveletTable", 
    current_plot="Mesh", 
    x='x', 
    scale="c",
    series='y', 
    vmin=-0.5,
    vmax=0.5,
    cmap="ice",
    title='Mesh Plot Example',
    heading= 'Simple Mesh Plot')
```

<hr style="height:2px;border:none;background-color:#00233C;">

#### Example 6 - Wiggle Plot
Below is an example of a Wiggle Plot, we use "waveletTable" as the input table.

- In the _Chart Tab_, set **X** to "x", **Scale** to "c", and **Series[0]** to "y".
- In the _Format Tab_, set **Title** to “Wiggle Plot Example” and **Heading** to “Simple Wiggle Plot”

#### Example 7 - Geometry Plot

Below is an example of a Geometry Plot, below loads the table and join shapes with population and filter only 1990 data.



```python
from teradataml import GeoDataFrame, load_example_data, DataFrame
load_example_data("geodataframe", ["sample_shapes"])
shapes_df = GeoDataFrame("sample_shapes")
load_example_data("geodataframe", ["us_population", "us_states_shapes"])
us_population = DataFrame("us_population")
# Join shapes with population and filter only 1990 data.
us_states_shapes = GeoDataFrame("us_states_shapes")
population_data = us_states_shapes.join(us_population,
                                         on=us_population.state_name == us_states_shapes.state_name,
                                         lsuffix="us",
                                         rsuffix="t2")
population_data_1990 = population_data[population_data.population_year == 1990]
population_data_1990
```


- In the _Chart Tab_, set **X** to "population_year" and **Series[0]** to "population".
- In the _Format Tab_, set **Title** to "Population Density across the US in 1990", set **Heading** to “Map Plot Example”, set **Color Map** to "rainbow", and **Width** to 800.
- In the _Range Tab_, set **Reverse Y Ticks** to True.

```python
from teradatamlwidgets import plot

plot7 = plot.Ui(
    df=population_data_1990,
    current_plot="Geom", 
    series='population', 
    title='Population Density across the US in 1990',
    reverse_yaxis= True,
    cmap='rainbow',
    width='800',
    heading='Map Plot Example')
```

## Subplot

Below is the code to have subplots i.e. show multiple plots together. This is a way of showing the plots created previously. It uses the function .ShowPlots() to do this.

ShowPlots takes in the plots wanting to be displayed in a list, and then has nrows for how many rows and ncols for how many columns. It also has a grid parameter which lets you specify which positioning you desire for the plots.

```python
from teradatamlwidgets import plot
plot.ShowPlots([plot1, plot2, plot3], nrows=1, ncols=3)
```

```python
plot.ShowPlots([plot1, plot2, plot3], grid = {(1, 1): (1, 1), (1, 2): (1, 1), (2, 1): (1, 2)})
```
