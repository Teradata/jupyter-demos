<b style = 'font-size:20px;font-family:Arial;'>Authenticate into User Environment Service (UES)</b>
<p style = 'font-size:16px;font-family:Arial;'>UES authentication is required to create and manage the Python or R environments that we will be creating. A VantageCloud user can easily create the authentication objects using the Console in a VantageCloud environment. The step to create these authentication objects has already been performed for you.</p>

```python
# We've already loaded all the values into our environment variables and into a dictionary, env_vars.
# username=env_vars.get("username") isn't required when using base_url, pat and pem.
ues_uri=env_vars.get("ues_uri")
if ues_uri.endswith("/open-analytics"):
    ues_uri = ues_uri[:-15]   # remove last 5 chars ("/open-analytics")

if set_auth_token(base_url=ues_uri,
                  pat_token=env_vars.get("access_token"), 
                  pem_file=env_vars.get("pem_file")
                 ):
    print("UES Authentication successful")
else:
    print("UES Authentication failed. Check credentials.")
    sys.exit(1)
```

<p style = 'font-size:18px;font-family:Arial;'> <b>Example of Vector Store use</b></p>
<p style = 'font-size:16px;font-family:Arial;'>This example will work through the whole UI as an example for how to use the UI.This assumes use of an AWS connection.<br>
    <b>For Creation: </b>   
<ul  style = 'font-size:16px;font-family:Arial;'>
    <li>Step 1:Click the Health heart button to see health status of the Vector Store</li>
    <li>Step 2:Click the Add button on left panel to create new vector table.In the Names section:<br>
        Set “Vector Store Name” to any name, for instance “vector_store_example”<br>
        Set “Object Names” to "vs_sample_data"<br>
        In the Columns section:Set “Data Columns” to comment_text, “key Columns” to comment_id</li>
    <li>Click the green Create button</li>
    <li>Click the information icon to the right of the Create button to see the status of the vector table, which should show as READY</li>
    <li>Step 3:Click the Refresh button to see the new vector table in list</li>
    <li>Step 4:Click the newly created vector table and click the Use check button</li>
    </ul>
       

<p style = 'font-size:16px;font-family:Arial;'><b>To Use and Ask questions: </b>   
<ul  style = 'font-size:16px;font-family:Arial;'>Select the newly created vector store (the check mark icon) to a ask questions 
    <li>For the Question enter “Which item has most comments?”<br>
Click check box for Similarity Search<br>
Click green Run button</li>
        <li>For the Question enter “Did any one say the about the material of shirt?”<br>
For the Prompt enter 'Format the response in a conversational way.'<br>
Click check box for Prepare Response (both Similarity Search and Prepare Response should be selected)<br>
 Click green Run button</li>
    <li>Click the Details button to see details of vector table</li>
    <li>Click Update button: Add Description as anything, for instance “This is an example for teradatamlwidgets Vector Store”. Click the green Update button</li>
       </ul>
