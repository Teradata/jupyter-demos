### Connect to Vantage

```python
import warnings
warnings.filterwarnings('ignore')
import time
import teradataml as tdml
tdml.display.print_sqlmr_query = False
import json
tdml.__version__
```

```python
with open('../connections/sasha.json', 'r') as f:
    Param_ = json.load(f)
    Param_['username'] = Param_['user']
    Param = {k:Param_[k] for k in ['host','username','password','database']}

tdml.create_context(**Param)
```

### Connect to the data

```python
dataset = tdml.DataFrame(tdml.in_schema('emails' , 'emails'))
dataset
```

### 4.2 - Compute Vector Embeddings

```python
from tdstone2.tdsgenai import compute_vector_embedding_byom
```

```python
res = compute_vector_embedding_byom(
    # choose your language model
    model              = 'tdstone2_emb_256_BAAI_bge-small-en-v1.5',
    # the description of the dataset
    dataset            = dataset,           # the teradata dataframeS
    text_column        = 'txt',   # the column containing the text
    accumulate_columns = ['id'],  # the columns we want to keep in the output results
    # the output table
    schema_name        = Param['database'], # the database 
    table_name         = 'embeddings',       # the table name (actually a view on a result table)
    primary_index      = ['id']            # the hash column to distribute the computations
)
```
