<b style = 'font-size:20px;font-family:Arial'>2. Data Exploration</b>
<p style = 'font-size:16px;font-family:Arial'>Create a "Virtual DataFrame" that points to the data set in Vantage. Check the shape of the dataframe as check the datatype of all the columns of the dataframe.</p>

```python
tdf = DataFrame(in_schema("DEMO_CreditCard","Credit_Card"))
print("Shape of the data: ", tdf.shape)
tdf
```

```python
tdf.tdtypes
```

```python
help(GetRowsWithoutMissingValues)
```

<p style = 'font-size:16px;font-family:Arial'>Let us check which columns have nulls in data.</p>

```python
colsum = ColumnSummary(data=tdf,
                             target_columns=[':']
                            )
 
cs = colsum.result.filter(items = ['ColumnName', 'Datatype', 'NonNullCount','NullCount','NullPercentage'])
cs[cs['NullPercentage'] > 0.0]
```

<p style = 'font-size:16px;font-family:Arial'>Now we get the rows which do not have NULLs for column OCCUPATION_TYPE.</p>

```python
 #Get the rows that contain NULL values in column OCCUPATION_TYPE
obj = GetRowsWithoutMissingValues(data=tdf,
                                target_columns=['OCCUPATION_TYPE'])
 
 # Print the result DataFrame.
obj.result
```

```python
obj.result.shape
```

<p style = 'font-size:16px;font-family:Arial'>From the results we can get the rows without null values in column OCCUPATION_TYPE in data.</p>
